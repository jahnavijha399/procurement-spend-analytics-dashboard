```python
import pandas as pd
import numpy as np
from faker import Faker
import random
```

```python
pip install faker
```

```python
fake = Faker()
```

```python
vendors = ["Dell","HP","Amazon","Cisco","Lenovo","Microsoft","Oracle","Infosys","TCS","Wipro"]
categories = ["IT Hardware", "Office Supplies", "Software", "Logistics", "Consulting"]
departments = ["IT","Finance","HR","Operations","Marketing"]
payment_status = ["Paid","Pending"]
regions = ["North","South","East","West"]
```

```python
data = []
```

```python
for i in range(10000):
    row = {
        "Invoice_ID":f"INV{i+1}",
        "Vendor_Name":random.choice(vendors),
        "Category": random.choice(categories),
        "Department":random.choice(departments),
        "Invoice_Date":fake.date_between(start_date="-2y",end_date="today"),
        "Amount": random.randint(500,50000),
        "Payment_status":random.choice(payment_status),
        "Region":random.choice(regions)
    }
    data.append(row)
```

```python
df=pd.DataFrame(data)
```

```python
df.to_csv("procurement_raw.csv", index=False)
```
